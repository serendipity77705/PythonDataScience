## Job Role and Workload Level Impact on Mental Health

Instructions: Investigate the influence of job roles and workload level on the mental health index.
- Create a new column `workload_level` that labels each entry as "High" if the `Hours_Worked_Per_Week` is above its mean, otherwise "Low." (5 marks)
- Group the dataset by `Industry` and calculate the average `Hours_Worked_Per_Week` for each combination. (5 marks)
- Use a bar chart to display the average `Stress_Level` for each job role, with separate bars for high and low stress levels. (5 marks)
- Analyze the results: Which job roles and workload levels appear to have the greatest impact on mental health? (5 marks)


```python
# Write code here
industry = df.groupby('Industry')
industry_mean_hours = industry['Hours_Worked_Per_Week'].mean()
mean_hours = df['Hours_Worked_Per_Week'].mean()

# breaking things into high and low stress levels
df['workload_level'] = np.where(df['Hours_Worked_Per_Week'] > mean_hours, 'High', 'Low')

high_stress = df[df['Stress_Level'] == 'High'].count()
low_stress = df[df['Stress_Level'] == 'Low'].count()
average_stress = df.groupby('Stress_Level')['Hours_Worked_Per_Week'].mean()

# grouping stress by job roles
high_stress_hours = high_stress.groupby('Job_Role')['Hours_Worked_Per_Week'].mean()
low_stress_hours = low_stress.groupby('Job_Role')['Hours_Worked_Per_Week'].mean()

# gives the job role labels
x = high_stress_hours.index

# used to separate high and low so they're not over each other
width = 0.25

plt.figure(figsize=(10,5))
plt.bar(x, high_stress_hours.values, width, color='red', label='High Stress')
plt.bar(np.arange(len(x)) + width, low_stress_hours.values, width, color='green', label='Low Stress')


plt.xlabel('Industry')
plt.ylabel('Average Hours Worked Per Week')
plt.title('Average Hours Worked Per Week by Industry Correlated to Stress Level')
plt.legend()
plt.show()

# the job that shows the largest difference between high and low stress levels is
# data scientist
```
